# Example 2: Modelstore with ScikitLearn

Creating a toy model for SVR with SKLearn, using model inspiration from [link](https://www.geeksforgeeks.org/support-vector-regression-svr-using-linear-and-non-linear-kernels-in-scikit-learn/)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
```

```python
# generate synthetic data
X = np.sort(5 * np.random.rand(40, 1),
            axis=0)
y = np.sin(X).ravel()
  
# add some noise to the data
y[::5] += 3 * (0.5 - np.random.rand(8))
```

```python
# create an SVR model with a linear kernel
svr = SVR(kernel='linear')
  
# train the model on the data
svr.fit(X, y)
  
# make predictions on the data
y_pred = svr.predict(X)
```

```python
# plot the predicted values against the true values
plt.scatter(X, y, color='darkorange',
            label='data')
plt.plot(X, y_pred, color='cornflowerblue',
         label='prediction')
plt.legend()
plt.show()
```

```python
# create an SVR model with a linear kernel
svrp = SVR(kernel='poly')
  
# train the model on the data
svrp.fit(X, y)
  
# make predictions on the data
y_pred = svrp.predict(X)
  
# plot the predicted values against the true values
plt.scatter(X, y, color='darkorange',
            label='data')
plt.plot(X, y_pred, color='cornflowerblue',
         label='prediction')
plt.legend()
plt.show()
```
